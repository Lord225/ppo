```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

# generate random 2d data and blur it
size = (8, 8)
original = np.random.random(size)

# scale
def blur(original):
    data = cv2.GaussianBlur(original, (3, 3), 0)
    data = cv2.resize(data, (0, 0), fx=4, fy=4, interpolation=cv2.INTER_LINEAR)
    return cv2.GaussianBlur(data, (5, 5), 0)

data = blur(original)

# plot
plt.imshow(data)
```

```python
# random variiation of original data
noise = np.random.normal(-0.3, 0.3, size)

# add noise to original data
base2 = original + noise

# blur
data2 = blur(base2)

# plot
plt.imshow(data2)
```
